```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow # For displaying images in Colab
# Upload an image
from google.colab import files
uploaded = files.upload()
# Read and display the image in color
image_path = list(uploaded.keys())[0] # Get the uploaded image file name
image = cv2.imread(image_path) # Load image in BGR format (default in OpenCV)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) # Convert to RGB for correct display in Matplotlib
plt.figure(figsize=(6, 6))
plt.imshow(image)
plt.title("Original Image")
plt.axis('off')
plt.show()
```

```python
# Invert the image (Negative transformation)
inverted_image = 255 - image
plt.figure(figsize=(6, 6))
plt.imshow(inverted_image, cmap='gray')
plt.title("Inverted Image")
plt.axis('off')
plt.show()
```

```python
# Contrast stretching using Min-Max Normalization
min_pixel = np.min(image)
max_pixel = np.max(image)
contrast_stretched = ((image - min_pixel) / (max_pixel - min_pixel) * 255).astype(np.uint8)
plt.figure(figsize=(6, 6))
plt.imshow(contrast_stretched, cmap='gray')
plt.title("Contrast Stretched Image")
plt.axis('off')
plt.show()
```

```python
# Apply histogram equalization
# Convert the image to grayscale first
gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
equalized_image = cv2.equalizeHist(gray_image) # Apply equalization to the grayscale image
plt.figure(figsize=(6, 6))
plt.imshow(equalized_image, cmap='gray')
plt.title("Histogram Equalized Image")
plt.axis('off')
plt.show()
```

```python
# Apply Canny edge detection
edges = cv2.Canny(image, 100, 200) # Adjust thresholds as needed
plt.figure(figsize=(6, 6))
plt.imshow(edges, cmap='gray')
plt.title("Edge Detected Image")
plt.axis('off')
plt.show()
```
